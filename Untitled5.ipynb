```python
# import the required libraries 
import random 
import matplotlib.pyplot as plt 
t=np.linspace(0, 960, num=961)
    
# store the random numbers in a  
# list 
nums = [] 
alpha = 100
beta = 0.5
    
for i in range(960): 
    temp = random.gammavariate(alpha, beta)
    nums.append(temp) 
        
# plotting a graph 
plt.plot(nums,) 
plt.show()
```

```python

# import the required libraries 
import random 
import matplotlib.pyplot as plt 
    
# store the random numbers in a list 
nums = [] 
alpha = 9
beta = 0.5
    
for i in range(10000): 
    temp = random.gammavariate(alpha, beta)
    nums.append(temp) 
        
# plotting a graph 
plt.hist(nums, bins = 200) 
plt.show()
```
